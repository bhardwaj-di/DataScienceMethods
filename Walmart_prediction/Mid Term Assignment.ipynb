1.Which store/s has good quarterly growth rate in Q3’2012?

```python
data['Date'] = pd.to_datetime(data['Date'])
Q3StartingDate = pd.Timestamp(date(2012,7,1))
Q3EndDate = pd.Timestamp(date(2012,9,30))
Q2StartingDate = pd.Timestamp(date(2012,4,1))
Q2EndDate = pd.Timestamp(date(2012,6,30))
```

```python
Q2data=data[(data['Date'] > Q2StartingDate) & (data['Date'] < Q2EndDate)]
Q3data=data[(data['Date'] > Q3StartingDate) & (data['Date'] < Q3EndDate)]
```

```python
Q2 = pd.DataFrame(Q2data.groupby('Store')['Weekly_Sales'].sum())
Q2.reset_index(inplace=True)
Q2.rename(columns={'Weekly_Sales': 'Q2_Weekly_Sales_Total'},inplace=True)

Q3 = pd.DataFrame(Q3data.groupby('Store')['Weekly_Sales'].sum())
Q3.reset_index(inplace=True)
Q3.rename(columns={'Weekly_Sales': 'Q3_Weekly_Sales_Total'},inplace=True)

Q3GrowthRate= Q2.merge(Q3,how='inner',on='Store')
```

```python
Q3GrowthRate['Growth_Rate'] =(Q3GrowthRate['Q3_Weekly_Sales_Total'] - Q3GrowthRate['Q2_Weekly_Sales_Total'])/Q3GrowthRate['Q2_Weekly_Sales_Total']
Q3GrowthRate['Growth_Rate']=round(Q3GrowthRate['Growth_Rate'],2)
Q3GrowthRate.sort_values('Growth_Rate',ascending=False).head(1)
```

4.Draw scatter plot of monthly sales and provide your observations

```python
date_2010_start = pd.Timestamp(date(2010,1,1))
date_2010_end = pd.Timestamp(date(2010,12,31))
data_2010=data[(data['Date'] > date_2010_start) & (data['Date'] < date_2010_end)]

date_2011_start = pd.Timestamp(date(2011,1,1))
date_2011_end = pd.Timestamp(date(2011,12,31))
data_2011=data[(data['Date'] > date_2011_start) & (data['Date'] < date_2011_end)]

date_2012_start = pd.Timestamp(date(2012,1,1))
date_2012_end = pd.Timestamp(date(2012,12,31))
data_2012=data[(data['Date'] > date_2012_start) & (data['Date'] < date_2012_end)]
```

```python
x=data_2010.groupby(pd.Grouper(key='Date', axis=0, 
                      freq='M'))['Weekly_Sales'].sum()
x=x.reset_index()

y=data_2011.groupby(pd.Grouper(key='Date', axis=0, 
                      freq='M'))['Weekly_Sales'].sum()
y=y.reset_index()

z=data_2012.groupby(pd.Grouper(key='Date', axis=0, 
                      freq='M'))['Weekly_Sales'].sum()
z=z.reset_index()
```

5.1 Can you predict Weekly Sales using other columns, either with a Random Forest Algorithm or a Support Vector Machine? 

```python
data=pd.read_csv("WALMART_SALES_DATA.csv")
data.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

sales_data=data
```

```python
sales_data = sales_data.drop('Date', axis=1)
```

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
sales_data = onehot_encode(sales_data, column='Store', prefix='store')
```

```python
sales_data['predict'] = sales_data['Weekly_Sales'].shift(-30)
sales_data
```

```python
sales_data2 = sales_data[0:-30]
sales_data2
```

```python
x = sales_data2['Weekly_Sales']
y = sales_data2['predict']
```
